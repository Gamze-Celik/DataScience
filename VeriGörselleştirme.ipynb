DİAMONDS VERİ SETİ:

VERININ TANIMLANMASI VE OZETLENMESI

PRICE : ABD Doları cinsinden fiyat ( 326 - 18.823 ABD Doları)

CARAT : Elmasın karat ağırlığı (0,2 - 5,01)

CUT : Kesim kalitesi (Adil, İyi, Çok İyi, Premium, Ideal)

COLOR : Elmas rengi, J (en kötü) ile D (en iyi)

CLARITY : Netliğin bir elmasın ne kadar net oldugunun ölçülmesi (I1 (en kötü), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (en iyi))

X : mm olarak x uzunluk (0–10.74)

Y : mm olarak y genişliği (0-58.9)

Z : mm cinsinden z derinliği (0-31.8)

DEPTH : Toplam derinlik yüzdesi = z / ortalama (x, y) = 2 * z / (x + y) (43 - 79)

TABLE : En geniş noktaya göre elmasın tablasının genişliği (43 - 95)

SÜTUN GRAFİK ÇARPRAZLAMALAR

```python
import seaborn as sns
from pandas.api.types import CategoricalDtype
diamonds=sns.load_dataset('diamonds')
df=diamonds.copy()
cut_kategoriler = ["Fair","Good","Very Good","Premium","Ideal"]
df.cut = df.cut.astype( CategoricalDtype(categories = cut_kategoriler, ordered = True ))
df.head()
```

```python
sns.catplot(x = "cut", y="price", data=df);
```

```python
sns.barplot(x ="cut", y ="price", hue="color", data = df);
```

```python
df.groupby(["cut","color"])["price"].mean()
```
